```python
import numpy as np
import random

# Step 1: Define the objective function
def objective_function(x):
    return x ** 2

# Step 2: Initialize Parameters
population_size = 20  # Number of candidate solutions
mutation_rate = 0.1   # Probability of mutation
crossover_rate = 0.7  # Probability of crossover
generations = 50      # Number of generations to run
bounds = (-10, 10)    # Range of possible values for x

# Step 3: Create Initial Population
population = np.random.uniform(bounds[0], bounds[1], population_size)

# Step 4: Evaluate Fitness
def evaluate_fitness(population):
    return np.array([objective_function(x) for x in population])

# Step 5: Selection
def selection(population, fitness):
    total_fitness = np.sum(fitness)
    selection_probabilities = fitness / total_fitness
    return np.random.choice(population, size=2, p=selection_probabilities)

# Step 6: Crossover
def crossover(parent1, parent2):
    if random.random() < crossover_rate:
        alpha = random.random()  # Random weight
        offspring1 = alpha * parent1 + (1 - alpha) * parent2
        offspring2 = (1 - alpha) * parent1 + alpha * parent2
        return offspring1, offspring2
    else:
        return parent1, parent2  # No crossover

# Step 7: Mutation
def mutate(individual):
    if random.random() < mutation_rate:
        return np.random.uniform(bounds[0], bounds[1])
    return individual

# Step 8: Run the Genetic Algorithm
best_solution = None
best_fitness = float('-inf')  # Start with the worst fitness

for generation in range(generations):
    fitness = evaluate_fitness(population)  # Evaluate fitness of the population
    new_population = []

    # Create a new population
    for _ in range(population_size // 2):  # Pair up for crossover
        parent1, parent2 = selection(population, fitness)  # Select parents
        offspring1, offspring2 = crossover(parent1, parent2)  # Create offspring
        new_population.append(mutate(offspring1))  # Mutate offspring
        new_population.append(mutate(offspring2))  # Mutate offspring

    population = np.array(new_population)  # Update population

    # Track the best solution found
    current_best_fitness = np.max(fitness)
    if current_best_fitness > best_fitness:
        best_fitness = current_best_fitness
        best_solution = population[np.argmax(fitness)]  # Update best solution

    # Print generation info
    print(f'Generation {generation + 1}: Best Fitness = {best_fitness}, Best Solution = {best_solution}')

# Output the best solution found
print(f'\nBest Solution Found: x = {best_solution}, f(x) = {best_fitness}')
```
